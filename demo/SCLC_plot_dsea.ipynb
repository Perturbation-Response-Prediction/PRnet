```python
cell_types = ['H446', 'H209', 'DMS114', 'H196', 'H526']
candi = ['(+)-Fangchinoline','(+)-JQ1','7257']
herb_fc_es_all = pd.DataFrame()
for cell_line in cell_types:
    candi_cell = [i+cell_line for i in candi]
    herb_fc_es=pd.read_csv(+cell_line+'fold-change_es.csv', index_col=0)
    herb_fc_es_all = pd.concat([herb_fc_es_all, herb_fc_es])
    
```

```python
fc_es_all_plot_1 = herb_fc_es_all.loc[:,['(+)-Fangchinoline','(+)-JQ1','7257']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)
# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(3, 13))
sns.heatmap(fc_es_all_plot_1, cmap=cmap, annot=True, fmt=".3f", linewidths=.5, ax=ax)
f.tight_layout()
f.subplots_adjust(left=0.1,right=0.9,top=0.9,bottom=0.1)
#f.savefig('./figures/SCLC_anti_cancer.pdf',format='pdf')
```
