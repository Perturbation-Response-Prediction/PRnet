```python
sc.pp.neighbors(latent_adata_plot)
sc.tl.tsne(latent_adata_plot)
sc.pl.tsne(latent_adata_plot, color=['cell_type'], wspace=0.4, frameon=False,save='LUNG_600_profile_latent.pdf')
```

```python
tsne_df['cell_type']=latent_adata_plot.obs.cell_type
```
